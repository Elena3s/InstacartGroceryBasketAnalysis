# 4. Identify most loyal customers:

## 4.1. Split the data into groups based on 'product_name'

```python
# Group df by 'product_name':

df.groupby('product_name')
```

## 4.2. Apply a function to each group separately - average orders per department

```python
# Find average number of orders for each department:

df.groupby('department_id').agg({'order_number': 'mean'})
```

## 4.3. Combine the results into a dataframe or alternative data structure or create a new column in the current dataframe - create 'max_order' column

```python
# a. Create 'max_order' column with maximum orders for each user:
    
ords_prods_merged['max_order'] = ords_prods_merged.groupby(['user_id'])['order_number'].transform(np.max)
```

## 4.5. Create a loyalty flag

```python
# b. Create a flag that assigns a “loyalty” label to a user ID based on its corresponding max order value:

ords_prods_merged.loc[ords_prods_merged['max_order'] > 40, 'loyalty_flag'] = 'Loyal customer'
```

```python
ords_prods_merged.loc[(ords_prods_merged['max_order'] <= 40) & (ords_prods_merged['max_order'] > 10), 'loyalty_flag'] = 'Regular customer'
```

```python
ords_prods_merged.loc[ords_prods_merged['max_order'] <= 10, 'loyalty_flag'] = 'New customer'
```

```python
ords_prods_merged['loyalty_flag'].value_counts(dropna = False)
```

```python
# c. Check that everything is interpreted correctly and that the right flags are assigned:

ords_prods_merged[['user_id', 'max_order', 'loyalty_flag', 'order_number']].head(60)
```

## 5.3. Create a spending flag (+ 'mean_price' column)

```python
# Item 6. Create a spending flag for each user based on the average price across all their orders: 
    # Criteria:
        # - If the mean of the prices of products purchased by a user is < 10, then flag them as a “Low spender.”
        # - If the mean of the prices of products purchased by a user is >= 10, then flag them as a “High spender.”
```

```python
    # a. Create 'mean_price' column with average price for each user:
    
ords_prods_merged['mean_price'] = ords_prods_merged.groupby(['user_id'])['prices'].transform(np.mean)
```

```python
        # Check the new 'mean-price' column:
    
ords_prods_merged.tail()
```

```python
    # b. Create a flag that assigns a spending label to a user ID based on the average price across all their orders:

ords_prods_merged.loc[ords_prods_merged['mean_price'] < 10, 'spending_flag'] = 'Low spender'
```

```python
ords_prods_merged.loc[ords_prods_merged['mean_price'] >= 10, 'spending_flag'] = 'High spender'
```

```python
ords_prods_merged['spending_flag'].value_counts(dropna = False)
```

```python
ords_prods_merged.describe().apply(lambda s: s.apply('{0:.5f}'.format))
```

## 5.4. Create a frequency flag (+ 'meadian_days_since' column)

```python
# Item 7. Create an order frequency flag (regularity of a user’s ordering behavior acc. to the median of “days_since_prior_order”. 
    # The criteria:
    # - If the median of “days_since_prior_order” is higher than 20, then the customer should be labeled a “Non-frequent customer.”
    # - If the median is higher than 10 and lower than or equal to 20, then - “Regular customer.”
    # - If the median is lower than or equal to 10, then - “Frequent customer.”
```

```python
    # a. Create 'median_days_since' with median days since the previous order for each customer:
    
ords_prods_merged['median_days_since'] = ords_prods_merged.groupby(['user_id'])['days_since_prior_order'].transform(np.median)
```

```python
ords_prods_merged.tail(10)
```

```python
    # b. Create a flag that assigns a frequency label to a user ID based on the regularity of a user’s ordering behavior:

ords_prods_merged.loc[ords_prods_merged['median_days_since'] > 20, 'frequency_flag'] = 'Non-frequent customer'
```

```python
ords_prods_merged.loc[(ords_prods_merged['median_days_since'] > 10) & (ords_prods_merged['median_days_since'] <= 20), 'frequency_flag'] = 'Regular customer'
```

```python
ords_prods_merged.loc[ords_prods_merged['median_days_since'] <= 10, 'frequency_flag'] = 'Frequent customer'
```

```python
ords_prods_merged['frequency_flag'].value_counts(dropna = False)
```
