# 3. Sorting products into price ranges

   ## 3.1. Sorting a subset

```python
# Create 1 mil row subset of:

df = ords_prods_merged[:1000000]
```

```python
# Check the shape of df:

df.shape
```

### 3.1.1. ... using user defined function

```python
# Define a fuction:

def price_label(row):
    
    if row['prices'] <= 5:
        return 'Low range product'
    elif (row['prices'] > 5) and (row['prices'] <= 15):
        return 'Mid rage product'
    elif row['prices'] > 15:
        return 'High range product'
    else: return np.nan
```

```python
# Apply the function in a new column 'price_range':

df['price_range'] = df.apply(price_label, axis = 1)
```

```python
    # Check for results of the 'price_rage' flag:

    df['price_range'].value_counts(dropna = False)
```

```python
    # Check for the highest price in the subset:

    df['prices'].max()
```

### 3.1.2. ...using df.loc( )

```python
df.loc[df['prices'] > 15, 'price_range_loc'] = 'High range product'
```

```python
df.loc[(df['prices'] <= 15) & (df['prices'] > 5), 'price_range_loc'] = 'Mid range product'
```

```python
df.loc[df['prices'] <= 5, 'price_range_loc'] = 'Low range product'
```

```python
    # Count up the number of products within each label in the “price_range_loc” column, missing values included:

    df['price_range_loc'].value_counts(dropna = False)
```

# 4. Creating 'busiest_day' column

```python
# Find out on which day most orders take place:

ords_prods_merged['orders_day_of_week'].value_counts(dropna = False)
```

```python
# Create a list that will contain values: “Busiest day,” “Least busy,” and “Regularly busy”:

result = []

for value in ords_prods_merged['orders_day_of_week']:
    
    if value == 0:
        result.append('Busiest day')
    elif value == 4:
        result.append('Least busy')
    else:
        result.append('Regularly busy')
```

# 5. Creating 'busy_slow_days' column (2 busiest and 2 slowest days)

```python
# Find out how many orders take place on which day:

ords_prods_merged['orders_day_of_week'].value_counts(dropna = False)
```

#### - Days 0 and 1 are the two busiest days, and days 3 and 4 are the two slowest ones

```python
# Create a list that contains values: 'Busiest days', 'Slowest days' and 'Regularly busy':

busyslow = []

for x in ords_prods_merged['orders_day_of_week']:
    
    if x == 0:
        busyslow.append('Busiest days')
    elif x == 1:
        busyslow.append('Busiest days')
    elif x == 3:
        busyslow.append('Slowest days')
    elif x == 4:
        busyslow.append('Slowest days')
    else:
        busyslow.append('Regularly busy')
```

```python
busyslow
```

```python
# Create a new column 'busy_slow_days' and populate it with 'busyslow' list:

ords_prods_merged['busy_slow_days'] = busyslow
```

```python
# Item 3. Check values of 'busy_slow_days'

ords_prods_merged['busy_slow_days'].value_counts(dropna = False)
```

# 6. Creating 'busiest_period_of_day' column (with periods of time: 'Most orders', 'Average orders', and 'Fewest orders')

```python
# Find out the busiest, quietest and average hours:

ords_prods_merged['time_of_order, hour'].value_counts(dropna = False)
```

#### - Hours 9 through 16 are the busiest, hours 23 through 6 are the quietest, the rest are average.

```python
# Create a list that contains values: 'Most orders', 'Average orders', and 'Fewest orders'

busy_hrs = []

for y in ords_prods_merged['time_of_order, hour']:
    
    if y == 9 or y == 10 or y == 11 or y == 12 or y == 13 or y == 14 or y == 15 or y ==16:
        busy_hrs.append('Most orders')
    elif y == 23 or y == 0 or y == 1 or y == 2 or y == 3 or y == 4 or y == 5 or y == 6:
        busy_hrs.append('Fewest orders')
    else:
        busy_hrs.append('Average orders')
```

```python
# Create a new column 'busiest_period_of_day' and populate it with 'busyslow' list:

ords_prods_merged['busiest_period_of_day'] = busy_hrs
```

```python
# Item 5. Check values of 'busiest_period_of_day'

ords_prods_merged['busiest_period_of_day'].value_counts(dropna = False)
```
